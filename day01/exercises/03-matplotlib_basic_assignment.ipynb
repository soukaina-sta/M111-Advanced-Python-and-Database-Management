## Exercise 2
**Get total profit of all months and show line plot with the following Style properties**
- Line Style dotted and Line-color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.
- Line marker color as read
- Line width should be 3

**Expected output**

## Exercise 5
**Read face cream and facewash product sales data and show it using the bar chart**
The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

**Expected output**

<img src="../data/Facewash_and_facecream_sales_data.png" alt="sales_data" style="width: 700px;"/>


```python
# write your code here
x = data['month_number']
y1 = data['facecream']
y2 = data['facewash']
width =0.25
plt.figure(figsize=(9, 6))
p1 = plt.bar(x,y1, width = width,label ='face cream sales data')
p2 = plt.bar(x+width,y2, width =width , label='face wash sales data')
plt.xlabel('Month number')
plt.ylabel('Sales units in number')
plt.title('Facecream and facewash sales data')
plt.legend((p1[0], p2[0]), ('face cream sales data', 'face wash sales data'))
plt.grid(linestyle = '--')
plt.show()
```

## Exercise 8
**Calculate total sale data for last year for each product and show it using a Pie chart**

*Note: In Pie chart display Number of units sold per year for each product in percentage.*

**Expected output**

<img src="../data/sales_data_pi_chart.png" alt="sales_data_pi_chart" style="width: 700px;"/>


```python
# write your code here
d = []
d.append(data.facecream.sum())
d.append(data.facewash.sum())
d.append(data.toothpaste.sum())
d.append(data.bathingsoap.sum())
d.append(data.shampoo.sum())
d.append(data.moisturizer.sum())
mylabels = ["facecream", "facewash", "toothpaste", "bathingsoap","shampoo","moisturizer"]
plt.pie(d, labels = mylabels ,autopct='%1.1f%%', startangle=90)
#plt.legend(loc='right')
plt.show() 
```

## Exercise 9
**Read Bathing soap facewash of all months and display it using the Subplot**

**Expected output**

<img src="../data/Sales_data_of_a_facewash.png" alt="Sales_data_of_a_facewash" style="width: 700px;"/>


```python
# write your code here
x = data['month_number']
y1 = data.bathingsoap
y2 = data.facewash

fig, (ax1, ax2) = plt.subplots(2, sharex=True)
plt.subplot(2, 1,1)
plt.ylabel('Sales units in number')
plt.title('Bathingsoap sales data')
ax1.plot(x,y1,marker= 'o',c='k')
plt.subplot(2, 1, 2)
plt.xlabel('Month number')
plt.ylabel('Sales units in number')
plt.title('Facewash sales data')
ax2.plot(x,y2,marker= 'o',c='r')
plt.show()
```

## Exercise 10
**Read all product sales data and show it using the stack plot**

**Expected output**

<img src="../data/product_sales_stack_plot.png" alt="product_sales_stack_plot" style="width: 700px;"/>


```python
plt.plot([],[],color='m', label='facecream', linewidth=5)
plt.plot([],[],color='c', label='facewash', linewidth=5)
plt.plot([],[],color='r', label='toothpaste', linewidth=5)
plt.plot([],[],color='k', label='bathingsoap', linewidth=5)
plt.plot([],[],color='g', label='shampoo', linewidth=5)
plt.plot([],[],color='y', label='moisturizer', linewidth=5)
plt.stackplot(data.month_number, data['facecream'],data['facewash'],data['toothpaste'],data['bathingsoap'],data['shampoo'],
              data['moisturizer'],colors = ['m','c','r','k','g','y'])
plt.xlabel('Month number')
plt.ylabel('Sales units in number')
plt.title('All product sales data using stack plot')
plt.legend(loc='upper left')
plt.show()
```
